## BertTokenizer

토크나이저로 pretrain된 BERT의 BertTokenizer를 갖고 옵니다. 여러 종류를 시도해보세요.

- bert-base-uncased : 108MB param, all lowercase
- bert-large-cased : 340MB param, both upper and lower
- bert-base-cased : 108MB param, multi language, both upper and lower


```python
tokenizer = BertTokenizer.from_pretrained('bert-base-cased') # 대소문자 구분해서 토큰화

# 기사 카테고리 매핑 딕셔너리 생성
labels = {'business':0,
          'entertainment':1,
          'sport':2,
          'tech':3,
          'politics':4
          }
```

## Dataset

```python
class Dataset(torch.utils.data.Dataset):

    def __init__(self, df):
        self.labels = [labels[label] for label in df['category']] # category 열의 문자열 라벨을 숫자로 변환
        # text 열을 BerTokenizer로 토큰화 (return_tensors="pt": 결과를 파이토치 텐서로 변환)
        self.texts = [tokenizer(text,
                               padding='max_length', max_length = 512, truncation=True,
                                return_tensors="pt") for text in df['text']]

    def classes(self):
        return self.labels # 전체 데이터셋의 라벨 리스트 반환

    def __len__(self):
        return len(self.labels) # 전체 데이터셋의 크기 반환

    def get_batch_labels(self, idx):
        return np.array(self.labels[idx]) # idx 위치의 라벨 하나를 넘파이 배열로 반환

    def get_batch_texts(self, idx):
        return self.texts[idx] # idx 위치의 텍스트 데이터 하나를 반환

    def __getitem__(self, idx): # idx 위치의 샘플(텍스트와 라벨)을 튜플로 반환

        batch_texts = self.get_batch_texts(idx)
        batch_y = self.get_batch_labels(idx)

        return batch_texts, batch_y
```

## Train & Evaluate BertClassifier

pretrain된 BertModel을 불러옵니다. 다른 간단한 층들도 같이 쌓아줍니다.

- bert-base-cased: 12-layer, 768-hidden, 12-self attention heads, 110M parameters. Trained on cased English text.


다른 종류들의 pretrianed model은 아래 링크에서 확인할 수 있습니다.

https://huggingface.co/transformers/v2.9.1/pretrained_models.html

```python
class BertClassifier(nn.Module):

    def __init__(self, dropout=0.5):

        super(BertClassifier, self).__init__()

        self.bert = BertModel.from_pretrained('bert-base-cased') # 사전 학습된 BERT Model 불러옴
        self.dropout = nn.Dropout(dropout)
        self.linear = nn.Linear(768, 5) # BERT의 임베딩: 768차원 -> 분류할 클래스의 개수: 5개
        self.relu = nn.ReLU()

    def forward(self, input_id, mask):
        # BERT 모델 실행 - 모든 토큰에 대한 벡터 & [CLS] 토큰의 임베딩 벡터 를 반환함
        _, pooled_output = self.bert(input_ids= input_id, attention_mask=mask,return_dict=False) # pooled_output: [CLS] 토큰의 임베딩 벡터(문장 전체의 의미를 요약/대표하는 벡터)
        dropout_output = self.dropout(pooled_output)
        linear_output = self.linear(dropout_output) # 5차원 벡터로 변환 -> 각 클래스별 점수
        final_layer = self.relu(linear_output)

        return final_layer
```

```python
def train(model, train_data, val_data, learning_rate, epochs):

    train, val = Dataset(train_data), Dataset(val_data)

    train_dataloader = torch.utils.data.DataLoader(train, batch_size=2, shuffle=True)
    val_dataloader = torch.utils.data.DataLoader(val, batch_size=2)

    use_cuda = torch.cuda.is_available()
    device = torch.device("cuda" if use_cuda else "cpu")

    criterion = nn.CrossEntropyLoss() # 다중 클래스 분류 문제에 적합한 손실함수인 CrossEntropyLoss 사용
    optimizer = Adam(model.parameters(), lr= learning_rate) # Adam optimizer 사용

    if use_cuda: # GPU가 있다면 -> 모델과 손실함수를 GPU에 올림

            model = model.cuda()
            criterion = criterion.cuda()

    for epoch_num in range(epochs):

            total_acc_train = 0 # 정확도
            total_loss_train = 0 # 손실 누적합

            for train_input, train_label in tqdm(train_dataloader):

                train_label = train_label.to(device)
                mask = train_input['attention_mask'].to(device) # 실제 문장 토큰이면 1, 패딩된 위치이면 0
                input_id = train_input['input_ids'].squeeze(1).to(device) # 문장을 토큰별 id의 시퀀스로 변환한 것

                output = model(input_id, mask) # BertClassifier의 forward()로 클래스별 점수 출력

                batch_loss = criterion(output, train_label.long()) # 배치별 손실 계산
                total_loss_train += batch_loss.item() # 누적

                acc = (output.argmax(dim=1) == train_label).sum().item() # output.argmax(dim=1)는 모델이 예측한 클래스의 인덱스 -> 정답인 train_label과 비교
                total_acc_train += acc

                # 역전파 및 가중치 업데이트
                model.zero_grad()
                batch_loss.backward()
                optimizer.step()

            # validation set을 위한 변수 초기화
            total_acc_val = 0
            total_loss_val = 0

            with torch.no_grad():

                for val_input, val_label in val_dataloader:

                    val_label = val_label.to(device)
                    mask = val_input['attention_mask'].to(device)
                    input_id = val_input['input_ids'].squeeze(1).to(device)

                    output = model(input_id, mask)

                    batch_loss = criterion(output, val_label.long())
                    total_loss_val += batch_loss.item()

                    acc = (output.argmax(dim=1) == val_label).sum().item()
                    total_acc_val += acc

            print(
                f'Epochs: {epoch_num + 1} | Train Loss: {total_loss_train / len(train_data): .3f} | Train Accuracy: {total_acc_train / len(train_data): .3f} | Val Loss: {total_loss_val / len(val_data): .3f} | Val Accuracy: {total_acc_val / len(val_data): .3f}')
```

```python
def evaluate(model, test_data):

    test = Dataset(test_data)

    test_dataloader = torch.utils.data.DataLoader(test, batch_size=2)

    use_cuda = torch.cuda.is_available()
    device = torch.device("cuda" if use_cuda else "cpu")

    if use_cuda:

        model = model.cuda()

    total_acc_test = 0
    with torch.no_grad():

        for test_input, test_label in test_dataloader:

              test_label = test_label.to(device)
              mask = test_input['attention_mask'].to(device)
              input_id = test_input['input_ids'].squeeze(1).to(device)

              output = model(input_id, mask)

              acc = (output.argmax(dim=1) == test_label).sum().item()
              total_acc_test += acc

    print(f'Test Accuracy: {total_acc_test / len(test_data): .3f}')
```

```python
np.random.seed(112)
df_train, df_val, df_test = np.split(df.sample(frac=1, random_state=42),
                                     [int(.8*len(df)), int(.9*len(df))])

print(len(df_train),len(df_val), len(df_test))
```

```python
EPOCHS = 2 #EPOCH 수 늘려보기!
model = BertClassifier()
LR = 1e-6

train(model, df_train, df_val, LR, EPOCHS)
```

```python
evaluate(model, df_test)
```
